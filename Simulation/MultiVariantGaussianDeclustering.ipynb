```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean = [0,0]
cov = [[1,0],[0,1]]
X = np.random.multivariate_normal(mean, cov, 2000)
plt.figure(dpi=140, figsize=(5,5))
plt.plot(X[:, 0], X[:, 1], "+")
```

```python
X1 = X[:1000, :] + 2
X2 = X[1000:, :] - 2
plt.figure(dpi=140, figsize=(5,5))
plt.plot(X1[:, 0], X1[:, 1], "r+")
plt.plot(X2[:, 0], X2[:, 1], "g+")
```
